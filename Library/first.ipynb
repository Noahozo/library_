```python
class Book:
    def __init__(self,name,year,author,review):
        self.name= name
        self.year= year
        self.author= author
        self.review= review

    def __str__(self):
        return '''書名：《%s》
                  年份： %s 
                  作者：%s 
                  心得：%s '''    %(self.name,self.year,self.author,self.review)
```
